# SiEPIC Library

A demonstration of all the components from the SiEPIC library that are implemented in simphony.

```python
import warnings
from pathlib import Path

import numpy as np
import skrf
from tabulate import tabulate

from simphony.utils import wl2freq


np.set_printoptions(linewidth=150)
np.set_printoptions(precision=3)
```

```python
from simphony.libraries.siepic import GratingCoupler


gc = GratingCoupler()

wl = np.linspace(1.42, 1.68, 1000) * 1e-6
f = wl2freq(wl)
s = gc.s_params(wl)

with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    ntwk = skrf.Network(f=f, s=s)
ntwk.plot_s_db()
```

```python
from simphony.libraries.siepic import YBranch


yb = YBranch()

wl = np.linspace(1.50, 1.6, 1000) * 1e-6
f = wl2freq(wl)
s = yb.s_params(wl)

with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    ntwk = skrf.Network(f=f, s=s)
ntwk.plot_s_db()
```
