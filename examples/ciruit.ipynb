```python
import simphony.libraries.siepic as siepic
import simphony.libraries.analytic as analytic

netlist={
    "instances": {
        "splitter":"ybranch",
        "combiner": "ybranch",  
        "top1": "waveguide",
        "top2": "waveguide",
        "bottom": "waveguide",
        "phase_modulator": "phase_modulator",
        "voltage_source": "voltage_source",

        # "prng": "prng",

        # "pm2": "phase_modulator",
        # "vs2": "voltage_source",

        # "y1": "ybranch",
        # "y2": "ybranch",
    },
    "connections": {
        "splitter,port_2": "top1,o0",
        "splitter,port_3":"bottom,o0",

        "top1,o1":"phase_modulator,o0",
        "phase_modulator,o1":"top2,o0",
        "top2,o1":"combiner,port_2",   
        "bottom,o1": "combiner,port_3",

        "voltage_source,e0":"phase_modulator,e0",
        
        
        # Multiple connections, same nodes
        # "y1,port 2": "y2,port 2",
        # "y1,port 3": "y2,port 3",

        # Invalid Connection
        # "vs2,e0":"pm2,o0",
    },
    "ports": {
        "in": "splitter,port_1",
        "out": "combiner,port_1",
    }
}

models={
    "ybranch": siepic.y_branch,
    # "ybranch": analytic.optical_s_parameter(siepic.y_branch),
    "waveguide": analytic.Waveguide,
    "phase_modulator": analytic.PhaseModulator,
    "voltage_source": analytic.VoltageSource,
    "prng": analytic.PRNG,
}

settings={ 
    # "splitter": {"bad_setting": 10},
    "top1": {"length": 5},
    "top2": {"length": 5},
    "bottom": {"length": 10},
}
```

```python
from simphony.circuit import Circuit
ckt = Circuit(netlist, models)
ckt.display(inline=True)
```

```python
from simphony.simulation import SParameterSimulation

SParameterSimulation(ckt)
```
